```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# Standardization centers all data around 0 and scales it based on its variance, 
# preventing high-magnitude features from disproportionately influencing the model's training process.
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Lets start with data collection and analyises 
datadf = pd.read_csv("diabetes.csv")

# check the first 5 rows
# datadf.head()
# we have pregnancis, Gluccose, BP, skinThicknesn, insulin, BMI, diabetesPedigreeFunction, Age, Outcome
# the outcome 0: non-diabetics 1: diabetics

# lets check if we have any empty values 
# datadf.isnull().sum()
# the data has no null values, 

# in head found out that differert values have a differet range, so need to standardise those

# datadf.shape
# we have 768 rows(samples) and 8 columns(features) + 1 (the result)

# datadf.describe()
# this gives variour data about the dataset, mean, count, std, min, max for all columns

# datadf["Outcome"].value_counts()
# this shows the distrubtion of the labels
# outcome:
# Outcome
# 0    500
# 1    268
# we can see we have a lot of non-diabetic than diabetic

# datadf.groupby("Outcome").mean()
# based on the values of Outcome column, the mean of all the columns with that value of Outome
```

```python
# now we are gonna separate data and labels
X = datadf.drop(columns="Outcome", axis=1) # all features except Outcome column
Y = datadf["Outcome"] # only the outcome column
# X.head()
# Y.head()
```

```python
# Now we are gonna standardise the data, bring everything to a common range
# but before that, its a good practice to split the data sets
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y, random_state=20)
# stratify to divide the data in proper proportion 
scaler = StandardScaler()
# creating an object of class StandardScaler()
scaler.fit(X_train) # this is just storing the mean and standard deviation for each feature, saved in the internla paremeter of the scaler object
# fit(X) doesn't change the values of X, we can change that with transform method which performs a technique called
#Z-score Normalization, here Z-score = (x- mean)/std, where x is values of features,
#note: NEVER PERFORM THIS WITH THE TEST DATA, YOU WILL CAUSE DATA LEAKAGE..
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train = X_train_scaled
X_test = X_test_scaled
```

```python
# Now lets start with traning the model, 
classifier = svm.SVC(kernel="linear")
```

```python
classifier.fit(X_train, Y_train)
# here we have to send both X and Y, and the amount of time needed to train is proportional to size of dataset
# here the values are of X and Y are being stored as parameters, which are used to find out optimal hyperplan(for this model)
# optimal hyperplan: the one that has the largest margin—the greatest distance to the nearest training data point of any class.
# The SVM algorithm doesn't use all the training data to define this boundary. Instead, it relies on a small subset of critical data points: the Support Vectors.
# It Stores the Hyperplane Parameters: It calculates and stores the mathematical definition of the hyperplane itself (its weight vector and bias term).
```

```python
# now we will try to predict the Y_test values with just X_test
# and also get accuracy score of that
X_test_prediction = classifier.predict(X_test)
# When you later call classifier.predict(X_test), the model takes a new data point from X_text and determines which side of the learned hyperplane it falls on
accuracy = accuracy_score(X_test_prediction, Y_test)
print(accuracy)
```

```python
# Now i want to take input from the user, their medical information and use the model to predict it with it
input_data = [6, 148, 72, 35, 0, 33.6, 0.627, 50]
# this is list type, we will change it to numpy array
data = np.array(input_data)
# before reshaping
# print(data.shape)
# op: (8,) .... this is an 1D array
data = data.reshape(1,-1)
# the data is 1D when you create a numpy array, but with reshape its now a 2D array, and the model wants the input data to be 2D
# after reshaping
# print(data.shape)
# op: (1, 8) ... this is an 2D array

# now we have to standardise this data too
data_std = scaler.transform(data)
# since this is same as the X(in range and mean and std and other stuff), we dont need to do fit again, just transform

prediction = classifier.predict(data_std)
if prediction[0] ==1:
    print("Congratulation you have diabetes")
elif prediction[0] == 0:
    print("Congratulation you dont have diabetes")
```
